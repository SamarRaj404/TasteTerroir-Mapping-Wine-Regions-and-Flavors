```python
cols=['country','description','points','price','taster_twitter_handle','taster_name']
indices=[0,5,35,100,500,499]
sample_reviews=Wine.loc[indices,cols]
sample_reviews
```

```python
us_wines=Wine[Wine.country=='US']
us_wines
```

```python
top_wines=Wine.loc[(Wine.country.isin(['US','Italy'])) & (Wine.points >=95)]
top_wines
```

```python
# What is the median of the points column in the reviews DataFrame?

median_points=Wine.points.median()
median_points
```

```python
# What countries are represented in the dataset? (Your answer should not include any duplicates.

countries=Wine.country.unique()
countries
```

```python
# How often does each country appear in the dataset? Create a Series reviews_per_country mapping countries to the count of reviews of wines from that country.
review_each_country=Wine.country.value_counts()
review_each_country.plot.bar(color=plt.cm.tab20(range(len(review_each_country))))
```

```python
centered_price=Wine.price - Wine.price.mean()
centered_price
```

```python
n_trop = Wine.description.map(lambda desc: "tropical" in desc).sum()
n_fruity = Wine.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])
descriptor_counts
```

```python
plt.title(' Bottle of Wine')
colors=['red','green']
descriptor_counts.plot.bar(color=colors)
```

```python
def stars(review):
    if review.country=='US':
        return 3
    elif review.points >=90 :
        return 3
    elif review.points >=85:
        return 2
    else:
        return 1
stars_rating=Wine.apply(stars, axis='columns')
stars_rating
```

```python
reviews_written=Wine.groupby('taster_twitter_handle').size()
reviews_written
```

```python
reviews_written.plot.bar(color=plt.cm.tab20(range(len(Wine))))
plt.show()
```

```python
best_rating_per_price=Wine.groupby('price')['points'].idxmax().head(30).sort_index()
best_rating_per_price.plot.bar(color=plt.cm.tab20(range(len(Wine))))
plt.show()
```

```python
# What are the minimum and maximum prices for each variety of wine? 
price_extremes=Wine.groupby('variety').price.agg(['min','max']).head(30)
price_extremes
```

```python
price_extremes.plot.bar()
```

```python
# What are the most expensive wine varieties?
sorted_varities=price_extremes.sort_values(by=['min', 'max'], ascending=False)
sorted_varities
```

```python
sorted_varities.plot.bar()
```

```python
# whose values is the average review score given out by that reviewer

reviewer_mean_rating = Wine.groupby('taster_name').points.mean()
reviewer_mean_rating.plot.pie()
plt.title('Average Review by Reviewer',c="black")
```

```python
colors=plt.cm.tab20(range(len(Wine)))
Wine['province'].value_counts().head(10).plot.bar(color=colors)
plt.show()
```

```python
#California produces almost a third of wines reviewed in Wine Magazine!

Wine['points'].value_counts().sort_index().plot.bar(color=colors)
plt.title('Distribution of Wine Points')
plt.xlabel('Points')
plt.ylabel('Count')
plt.show()
plt.show()
```

```python
Wine[Wine.price < 200]['price'].plot.hist()
```

```python
#we want wine price more than 1500 and points have greater than 85
Wine[(Wine.price > 1500) & (Wine.points > 85)].plot.bar(color=colors)
plt.title('Wine Price')
plt.ylabel('price')
plt.xlabel('Total Wine')
```

```python
#which will give you the number of purchases by each taster. The resulting bar plot will show the distribution of wine purchases across different tasters.

wine_purchase=Wine.groupby('taster_name')['winery'].count()
wine_purchase.plot.bar(color=plt.cm.tab20(range(len(Wine))))
plt.title('Wine Purchases by Taster')
plt.xlabel('Taster Name')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
taster_designition_name=Wine.groupby('taster_name')['designation'].sum()
taster_designition_name
```

```python
# Total unique designation of Taster
taster_designation_count = Wine.groupby('taster_name')['designation'].nunique()
plt.scatter(taster_designation_count.index, taster_designation_count.values, c=plt.cm.tab20(range(len(taster_designation_count))))
plt.title('Total Designations by Taster')
plt.xlabel('Taster Name')
plt.ylabel('Number of Unique Designations')
plt.show()
```
